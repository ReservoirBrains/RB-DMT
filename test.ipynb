```python
import pandas as pd
import os
import numpy as np

X = []
Y = []

all_targets = pd.read_csv('targets.csv')
target_var = "Gender"
target = pd.get_dummies(all_targets["Gender"]).astype(int)

for filename in os.listdir('schaefer200')[:10]:
    if filename.endswith('.csv'):
        ID = int(filename.split('_')[0])
        d = pd.read_csv(f'schaefer200/{filename}', header=None).to_numpy()
        X.append(d)
        Y.append(target[all_targets["Subject"] == ID].to_numpy().reshape(-1))

X = np.array(X).transpose(0, 2, 1)
Y = np.array(Y) 

print(X.shape, Y.shape)
```
