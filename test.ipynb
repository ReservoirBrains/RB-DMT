```python
import pandas as pd
import os
import numpy as np

NB_TRAIN = 10
NB_TEST = 10

X_TRAIN = []
Y_TRAIN = []
X_TEST = []
Y_TEST = []

all_targets = pd.read_csv('targets.csv')
target_var = "Gender"
target = pd.get_dummies(all_targets["Gender"]).astype(int)

filenames = os.listdir('schaefer200')
ids = np.unique([x.split("_")[0] for x in filenames])
np.random.seed(0)
np.random.shuffle(ids[NB_TRAIN: NB_TRAIN + NB_TEST])

if NB_TEST + NB_TRAIN > len(ids):
    raise Exception("Not enough samples.")

for _id in ids[:NB_TRAIN]:
    name1 = f"{_id}_REST1_parcellated_timeseries_schaefer200.csv"
    name2 = f"{_id}_REST2_parcellated_timeseries_schaefer200.csv"
    d1 = pd.read_csv(f'schaefer200/{name1}', header=None).to_numpy()
    d2 = pd.read_csv(f'schaefer200/{name2}', header=None).to_numpy()
    X_TRAIN.append(d1)
    X_TRAIN.append(d2)
    Y_TRAIN.append(target[all_targets["Subject"] == int(_id)].to_numpy().reshape(-1))
    Y_TRAIN.append(target[all_targets["Subject"] == int(_id)].to_numpy().reshape(-1))

for _id in ids[NB_TRAIN: NB_TRAIN + NB_TEST]:
    name1 = f"{_id}_REST1_parcellated_timeseries_schaefer200.csv"
    name2 = f"{_id}_REST2_parcellated_timeseries_schaefer200.csv"
    d1 = pd.read_csv(f'schaefer200/{name1}', header=None).to_numpy()
    d2 = pd.read_csv(f'schaefer200/{name2}', header=None).to_numpy()
    X_TEST.append(d1)
    X_TEST.append(d2)
    Y_TEST.append(target[all_targets["Subject"] == int(_id)].to_numpy().reshape(-1))
    Y_TEST.append(target[all_targets["Subject"] == int(_id)].to_numpy().reshape(-1))

X_TRAIN = np.array(X_TRAIN).transpose(0, 2, 1)
X_TEST = np.array(X_TEST).transpose(0, 2, 1)
Y_TRAIN = np.array(Y_TRAIN) 
Y_TEST = np.array(Y_TEST) 

print(X_TRAIN.shape, X_TEST.shape, Y_TRAIN.shape, Y_TEST.shape)
```
